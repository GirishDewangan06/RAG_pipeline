```python
import torch
from transformers import pipeline
from sentence_transformers import SentenceTransformer, util


```

```python
# Load similarity & entailment models
similarity_model = SentenceTransformer('all-MiniLM-L6-v2')
entailment_model = pipeline("text-classification", model="facebook/bart-large-mnli")
```

```python
def compute_groundedness(statement, retrieved_docs):
    if not retrieved_docs:
        return 0.0  # No evidence, 0% grounded

    # Compute similarity between statement and retrieved knowledge
    statement_embedding = similarity_model.encode(statement, convert_to_tensor=True)
    doc_embeddings = similarity_model.encode(retrieved_docs, convert_to_tensor=True)
    similarities = util.pytorch_cos_sim(statement_embedding, doc_embeddings)

    # Maximum similarity score
    max_similarity = torch.max(similarities).item()
    SIMILARITY_THRESHOLD = 0.65
    similarity_score = min(max_similarity, 1.0)

    entailment_scores = []
    for doc in retrieved_docs:
        entailment_result = entailment_model([{ "text": statement, "text_pair": doc }])
        if entailment_result[0]['label'] == "ENTAILMENT":
            entailment_scores.append(entailment_result[0]['score'])


    max_entailment_score = max(entailment_scores) if entailment_scores else 0.0


    groundedness_score = max(similarity_score, max_entailment_score)
    hallucination_rate = 1 - groundedness_score

    return round(groundedness_score, 2), round(hallucination_rate, 2)
```

```python

def retrieve_ground_truth(user_query):
    knowledge_base = {
        "LDos attack": [
            """ LDoS attack detection and mitigation scheme. Our
                detection scheme collects the features of network traffic and uses
                the FSS-RSR algorithm to detect LDoS attacks in real time. Our
                mitigation scheme monitors traffic fluctuations to locate attack
                source IP addresses and delivers traffic rules to mitigate attacks."""
        ],

        "LDos detection":[
            """ The detection methods based on features analyze the differ
                ence in traffic features between the normal network and the
                network under attack and perform attack detection by extracting
                network features. This approach is often combined with mod
                els such as classification and clustering algorithms in machine
                learning"""
        ],
        "attack-source":[
            """ If an LDoS attack occurs, one or
                more attack sources exist. The LDoS attack path is as follows:
                attack source- link- victim. When LDoS attack traffic flows
                through the link switches, it will be recorded in the flow table.
                Therefore, the flow table can collect the traffic information of
                a single flow. The OpenFlow flow table is extracted at a fixed
                interval, and the Counter field of each flow table is counted
                to locate the attack traffic with periodic short-term pulses. The
                attack source can be located by querying the IP value in the
                matching domain of the flow table."""
        ]
    }
    for topic in knowledge_base:
        if topic.lower() in user_query.lower():
            return knowledge_base[topic]
    return []
```

```python
if isinstance(response, dict):
    response_text = response.get("output", "")
elif isinstance(response, list):
    response_text = " ".join(response)
else:
    response_text = str(response)

retrieved_docs = retrieve_ground_truth("LDos attack")
groundedness_score, hallucination_rate = compute_groundedness(response_text, retrieved_docs)

print("\nChatbot Response:", response_text)


print("Groundedness Score:", groundedness_score * 100, "%")
print("Hallucination Rate:", hallucination_rate * 100, "%")
```

```python
def compute_groundedness(statement, retrieved_docs):
    if not retrieved_docs:
        return 0.0, 1.0

    # Compute similarity between statement and retrieved knowledge
    statement_embedding = similarity_model.encode(statement, convert_to_tensor=True)
    doc_embeddings = similarity_model.encode(retrieved_docs, convert_to_tensor=True)
    similarities = util.pytorch_cos_sim(statement_embedding, doc_embeddings)

    max_similarity = torch.max(similarities).item() if similarities.numel() > 0 else 0.0
    similarity_score = min(max_similarity, 1.0)

    # Compute entailment using Groq's LLaMA model
    entailment_score = groq_entailment(statement, retrieved_docs)

    groundedness_score = max(similarity_score, entailment_score)
    hallucination_rate = 1 - groundedness_score

    return round(groundedness_score, 2), round(hallucination_rate, 2)
```
